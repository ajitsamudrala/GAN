```python
import os
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
IMAGE_DIR = Path('Data/celeb_data/')
```

```python
print(f"Number of image of celebs is {len(os.listdir(IMAGE_DIR))}")
```

```python
top_9_images = os.listdir(IMAGE_DIR)[:10]
top_9_images_paths = [IMAGE_DIR / image_name for image_name in top_9_images]
```

```python
images = [plt.imread(image_path) for image_path in top_9_images_paths]
images_shapes = [image.shape for image in images]
```

```python
images_shapes # all of them have same shape
```

```python
fig = plt.figure(figsize=(9, 9))
for i in range(9):
    ax = fig.add_subplot(3, 3, i + 1)
    ax.imshow(images[i], cmap='gray')
```
